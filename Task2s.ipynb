***MANIPULATING DATA:***

```python
#Group dataframe using a mapper or by a series of columns

import pandas as pd
data = {'City': ['New York', 'San Francisco', 'Los Angeles', 'New York', 'Los Angeles'],
        'Population': [8500000, 884000, 3970000, 8500000, 3970000],
        'Area': [468.9, 121.4, 468.9, 468.9, 468.9]}
df = pd.DataFrame(data)

# Group the DataFrame by the 'City' column
grouped_df = df.groupby('City')

# Perform an operation on each group, for example, calculate the mean
mean_population_by_city = grouped_df['Population'].mean()

# Display the mean population for each city
print("Mean Population by City:")
print(mean_population_by_city)

# Create a pivot table to calculate the mean population for each city
pivot_table_mean_population = df.pivot_table(values='Population', index='City', aggfunc='mean')

# Display the pivot table
print("Pivot Table - Mean Population by City:")
print(pivot_table_mean_population)
```

```python
#Sort dataframes by one or more columns


import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [25, 30, 35, 28, 22],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago', 'Miami']}
df = pd.DataFrame(data)

# Sort the DataFrame by the 'Age' column in ascending order
sorted_df_age_asc = df.sort_values(by='Age')

# Display the sorted DataFrame
print("Sorted DataFrame by Age (Ascending):")
print(sorted_df_age_asc)

# Sort the DataFrame by the 'Age' column in descending order
sorted_df_age_desc = df.sort_values(by='Age', ascending=False)

# Display the sorted DataFrame
print("\nSorted DataFrame by Age (Descending):")
print(sorted_df_age_desc)
```

COMBINING DATA FRAMES:

```python
import pandas as pd

# Example DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# Concatenate along rows (axis=0)
result = pd.concat([df1, df2])
print("Concatenated along rows:")
print(result)

# Concatenate along columns (axis=1)
result = pd.concat([df1, df2], axis=1)
print("\nConcatenated along columns:")
print(result)
```

```python
import pandas as pd

# Example DataFrames
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value': [1, 2, 3, 4]})

df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'],
                    'value': [5, 6, 7, 8]})

# Merge DataFrames based on the 'key' column
result_inner = pd.merge(df1, df2, on='key', how='inner')
print("Inner Merge:")
print(result_inner)

result_outer = pd.merge(df1, df2, on='key', how='outer')
print("\nOuter Merge:")
print(result_outer)

result_left = pd.merge(df1, df2, on='key', how='left')
print("\nLeft Merge:")
print(result_left)

result_right = pd.merge(df1, df2, on='key', how='right')
print("\nRight Merge:")
print(result_right)
```

***CHANGING DATA:***

```python
#Apply a function along the axis of the DataFrame.
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

# Define a function to square each element
def square_element(x):
    return x ** 2

# Apply the function element-wise to each column
result = df.apply(square_element)
print("Result of applying the function to each column:")
print(result)

# Apply the function element-wise to each row
result_row = df.apply(square_element, axis=1)
print("\nResult of applying the function to each row:")
print(result_row)
```

```python
#Map values of a Series using a lookup or a function.
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'A': ['apple', 'banana', 'cherry'],
    'B': ['orange', 'banana', 'apple']
})

# Define a mapping dictionary
fruit_mapping = {'apple': 'red', 'banana': 'yellow', 'cherry': 'red', 'orange': 'orange'}

# Use applymap to substitute values in the DataFrame
result = df.applymap(lambda x: fruit_mapping.get(x, x))
print("Result of mapping values in the DataFrame:")
print(result)
```

```python
#Resample time-series data.
import pandas as pd
import numpy as np

# Create a DataFrame with a datetime index
date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
df = pd.DataFrame({'value': np.arange(10)}, index=date_rng)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Resample the DataFrame to a weekly frequency, taking the mean of values within each week
weekly_resampled = df.resample('W').mean()

# Display the resampled DataFrame
print("\nResampled DataFrame (weekly mean):")
print(weekly_resampled)
```

```python
#Shift index by desired number of periods.
import pandas as pd
import numpy as np

# Create a DataFrame with a datetime index
date_rng = pd.date_range(start='2023-01-01', end='2023-01-05')
df = pd.DataFrame({'value': np.arange(5)}, index=date_rng)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Shift the values of the DataFrame by 1 period (one day in this case)
df_shifted = df.shift(1)

# Display the DataFrame with shifted values
print("\nDataFrame with Shifted Values (1 day lag):")
print(df_shifted)
```

***STATISTICAL FUNCTIONS:***

```python
import pandas as pd
import numpy as np

# Create a DataFrame
data = {'A': [1, 2, 3, 4, 5],
        'B': [5, 4, 3, 2, 1],
        'C': [2, 2, 2, 2, 2]}
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Calculate mean, median, and standard deviation
mean_values = df.mean()
median_values = df.median()
std_values = df.std()

# Display the calculated statistics
print("\nMean values:")
print(mean_values)

print("\nMedian values:")
print(median_values)

print("\nStandard deviation values:")
print(std_values)
```

```python
#Compute pairwise correlation.
import pandas as pd

# Create a DataFrame
data = {'A': [1, 2, 3, 4, 5],
        'B': [5, 4, 3, 2, 1],
        'C': [2, 2, 2, 2, 2]}
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Display the correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
#Drop specified labels from rows or columns.
import pandas as pd

# Create a DataFrame
data = {'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]}
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Drop a specific column (axis=1)
df_dropped_column = df.drop('B', axis=1)

# Display the DataFrame after dropping the column
print("\nDataFrame after dropping column 'B':")
print(df_dropped_column)

# Drop a specific row (axis=0)
df_dropped_row = df.drop(1, axis=0)

# Display the DataFrame after dropping the row
print("\nDataFrame after dropping row with index 1:")
print(df_dropped_row)





# Drop multiple columns
df_dropped_columns = df.drop(['A', 'C'], axis=1)

# Display the DataFrame after dropping columns 'A' and 'C'
print("\nDataFrame after dropping columns 'A' and 'C':")
print(df_dropped_columns)

# Drop multiple rows
df_dropped_rows = df.drop([0, 2], axis=0)

# Display the DataFrame after dropping rows with indices 0 and 2
print("\nDataFrame after dropping rows with indices 0 and 2:")
print(df_dropped_rows)

```

```python
import pandas as pd

# Create a DataFrame
data = {'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]}
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Save the DataFrame to a CSV file
df.to_csv('output.csv', index=False)  # index=False avoids writing row indices to the CSV file

# Read the CSV file back into a DataFrame
df_from_csv = pd.read_csv('output.csv')

# Display the DataFrame read from the CSV file
print("\nDataFrame read from the CSV file:")
print(df_from_csv)

# Save the DataFrame to a CSV file with a tab delimiter and UTF-8 encoding
df.to_csv('output_tab_delimited.csv', sep='\t', encoding='utf-8', index=False)
```
